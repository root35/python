# **ARRAY DATA STRUCTURES**
* Arrays = fixed-size data records
* allow each element to be efficiently located based on its index
* _contiguous_ data structures: store eleemnts in ajoining blocks of memory  
(as opposed to _linked_ data structures)
* Performance:
  * very fast lookup given the index
  * proper implementation guarantees __O(1)__ access time
