# **FUNCTION ARGUMENT UNPACKING**

```
def foo(a, b, c):
  return '<%s, %s, %s>' % (a, b, c)

l = ['a', 'b', 'c']
print('unpacking a list: ', foo(*l))

t = ('x', 'y', 'z')
print('unpacking a tuple: ', foo(*t))

g = (x for x in range(3))
print('unpacking a generator expression: ', foo(*g))

d = {'a': 1, 'b': 2, 'c': 3}
print('unpacking dictionary keys: ', foo(*d))
print('unpacking dictionary values: ', foo(**d))

d = {'x': 1, 'y': 2, 'z': 3}
foo(*d)
foo(**d)
# Will not work because the keys don't have the same names as the function's arguments
```
