# **IMPLICIT RETURNS**
* Only use them in procedures: functions with no `return` value at all.   
* Otherwise, it's better to use an explicit `return` statement: by clearly showing what the function is doing:
  * you will avoid any possible misunderstanding: some people might not know about this feature
  * and it will simplify the code maintenance: you won't have to explain implicit returns to them

```
def func1(value):
  if (value):
    return value
  else:
    return None # explicit return statement
```

```
def func2(value):
  if (value):
    return value
  else:
    return # 'return' is an implicit 'return None'
```

```
def func3(value):
  if (value):
    return value
  # No 'return' statement is an implicit 'return None'
```

```
print('func1: ', type(func1(False)), ' - ', type(func1(True)))
print('func2: ', type(func2(False)), ' - ', type(func2(23)))
print('func3: ', type(func3(False)), ' - ', type(func3('OK')))
```
