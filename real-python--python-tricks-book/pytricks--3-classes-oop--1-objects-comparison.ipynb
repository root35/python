# **COMPARING OBJECTS: `is` vs `==`**
* **`is`**: 
identical, same entity (same object)
* **`==`**: equal, 2 similar entities (same content)

```
a = [1, 2, 3]
b = a
print('a == b: ', a == b)
print('a is b: ', a is b)
```

```
a = [1, 2, 3]
b = list(a)
print('a == b: ', a == b)
print('a is b: ', a is b)
```
