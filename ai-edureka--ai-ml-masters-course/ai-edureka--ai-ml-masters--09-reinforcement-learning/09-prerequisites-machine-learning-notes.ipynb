<H3 style="color: blue; font-size: 15px; background-color: #ffffff; padding: 15px 20px;">Exploratory Analysis</H3>

Summarize main characteristics, often with visual representations

1. Gain maximum insight into the dataset and its underlying structure
2. Find relations between attributes
3. Check for missing data and other mistakes
4. List outliers and other anomalies

<a name="01-01"></a>
<H2 style="color: MediumBlue; font-size: 21px; background-color: #F0F8FF; padding: 15px 20px;">Measures of center</H2>

<H3 style="color: blue; font-size: 15px; background-color: #ffffff; padding: 15px 20px;">Mean</H3>

Average 
  
1. Add data values
2. Divide by the sample size

<H3 style="color: blue; font-size: 15px; background-color: #ffffff; padding: 15px 20px;">Median</H3>

Value in the middle
  
1. Rank data in ascending order
2. Find the number in the middle
  
    
* Uneven nb of values: 2 2 2 2 3 4 5 __5__ 6 7 8 8 9 **->** median is 5
* Even nb of values: 2 2 2 3 4 5 5 6 7 8 8 9 **->** median is the mean of the 2 middle values: (5 + 5) / 2 = 5

<H3 style="color: blue; font-size: 15px; background-color: #ffffff; padding: 15px 20px;">Mode</H3>

Most frequent value

1. Rank data in ascending order
2. Count frequency of each value
3. Value with highest frequency is the mode
   
* If all same frequency, then no mode
* If multiple values with highest frequency, then multiple modes -> **bi-modal, multi-modal**

<a name="01-01"></a>
<H2 style="color: MediumBlue; font-size: 21px; background-color: #F0F8FF; padding: 15px 20px;">Measures of spread, or dispersion</H2>

* Variability in the sample
* Used in cunjunction with a measure of center for overall description of the data.

<H3 style="color: blue; font-size: 15px; background-color: #ffffff; padding: 15px 20px;">Range</H3>

```
Range = Max value - Min value
```
Informs if at least one value broke a threshold 

<H3 style="color: blue; font-size: 15px; background-color: #ffffff; padding: 15px 20px;">Quartiles</H3>

Breaking the dataset into quarters:   
*just like the median breaks it in half*   
```
Q1 = mean(last of 1st quarter + first of 2d quarter)   
Q2 = mean(last of 2d quarter + first of 3d quarter)   
Q3 = mean(last of 3d quarter + first of 4th quarter) 
```

<H3 style="color: blue; font-size: 15px; background-color: #ffffff; padding: 15px 20px;">Inter-quartile range</H3>
  
```
IQR = Q3 - Q1
```

<H3 style="color: blue; font-size: 15px; background-color: #ffffff; padding: 15px 20px;">Variance</H3>

* How far each number is from the mean
* Calculation:
    1. take the difference between each number in the set and the mean
    2. square all the differences
    3. divide the sum of squares by the number of values in the dataset

<H3 style="color: blue; font-size: 15px; background-color: #ffffff; padding: 15px 20px;">Standard Deviation</H3>

Measure of dispersion from the mean

```
Std. Dev. = square root of the variance
```
