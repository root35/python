```python
import tkinter as tk
from tkinter import messagebox

class TicTacToe:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Tic Tac Toe")
        self.board = [["" for _ in range(3)] for _ in range(3)]
        self.current_turn = "X"

        self.buttons = [[tk.Button(self.window, text="", command=lambda row=row, col=col: self.click(row, col), height=3, width=6) for col in range(3)] for row in range(3)]
        for row in range(3):
            for col in range(3):
                self.buttons[row][col].grid(row=row, column=col)

    def click(self, row, col):
        if self.board[row][col] == "" and not self.check_win():
            self.buttons[row][col]['text'] = self.current_turn
            self.board[row][col] = self.current_turn
            if self.check_win():
                messagebox.showinfo("Game Over", f"{self.current_turn} wins!")
            elif "" not in [self.board[i][j] for i in range(3) for j in range(3)]:
                messagebox.showinfo("Game Over", "It's a draw!")
            else:
                self.current_turn = "O" if self.current_turn == "X" else "X"

    def check_win(self):
        for row in self.board:
            if row.count(row[0]) == len(row) and row[0] != "":
                return True

        for col in range(len(self.board)):
            check = []
            for row in self.board:
                check.append(row[col])
            if check.count(check[0]) == len(check) and check[0] != "":
                return True

        if self.board[0][0] == self.board[1][1] == self.board[2][2] and self.board[0][0] != "":
            return True
        if self.board[0][2] == self.board[1][1] == self.board[2][0] and self.board[0][2] != "":
            return True

        return False

    def run(self):
        self.window.mainloop()

if __name__ == "__main__":
    game = TicTacToe()
    game.run()
```
