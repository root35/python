```python
import numpy as np

def sigmoid_(a):
    return 1 / (1 + np.exp(-a))
x = [1, 2, 4, 0, 5, -6, 7]
for a in x:
    print(sigmoid_(a))
    
# If the value is bigger, it will be closer to 1.
```

```python
tanh(a) = (e^a - e^-a) / e^a + e^-a
```

```python
def tanh_(a):
    return np.tanh(a)
x = [-4, -2, 0, 2, 4]
for a in x:
    print(tanh_(a))
```

```python
def relu_(a):
    return np.maximum(0, a)  # relu: returns 0 if a < 0
x = [-4, -2, 2, 4]
for a in x:
    print(relu_(a))
```

```python
import numpy as np
scores = [12, 8, .3]
def softmax_(a):
    return np.exp(a) / np.sum(np.exp(a), axis = 0)
print(softmax_(scores))
print(sum(softmax_(scores)))
```

```python
import numpy as np
a = [.2, 1, 0, 3, -2]
sigmoid_activation = tf.sigmoid(a)
tanh_activation = tf.tanh(a)
relu_activation = tf.nn.relu(a)
```
