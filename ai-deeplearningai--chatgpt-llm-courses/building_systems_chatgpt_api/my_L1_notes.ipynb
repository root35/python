```python
import os
import openai
import tiktoken
from dotenv import load_dotenv
```

```python
# read local .env file: /Users/macbook/.env
# first activate py39 environment
_ = load_dotenv('/Users/macbook/.env')  
openai.api_key = os.environ['OPENAI_API_KEY']
```

```python
messages = [
    {'role': 'system',   # specify the overall tone
     'content': 'You are an assistant who writes in the style of Dr Seuss.'},
        # or "All your responses must be 1 sentence long."
    {'role': 'user',     # instruction to carry out
     'content': 'Tell me a joke about carrots.'},
    {'role': 'assistant',
     'content': 'You are my assistant.'},
]
```

<img src="images/008.png" width="350" style="border: 1px solid black;">

```python
def get_completion_from_messages(messages, 
                                 model="gpt-3.5-turbo", 
                                 temperature=0, 
                                 max_tokens=500):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, 
        max_tokens=max_tokens,
    )
    return response.choices[0].message["content"]
```

```python
response, token_dict = get_completion(messages,
                          "What is the capital of Morocco?",
                          temperature=1)
print(response)
print(token_dict)
```

```python
response = get_completion("Reverse the letters in the word lollipop")
print(response)

# fails because not frequent word > tokenization not correct 'l', 'oll', 'ipop'
# possible solution: dashes
#  > takes each letter as a single token, so easier to reverse
response = get_completion("Reverse the letters in the word l-o-l-l-i-p-o-p")
print(response)
```
