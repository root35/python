# Sampling
Inferential statistics
* How many is enough?
* Best samples are chosen at random
* Alternatives:
  * 1 person every k
  * first n units
  * stratified samples: sample of each category
  * cluster sample: sample of each random cluster
* Larger sample = greater accuracy (higher chance of representing the total population)
* **Central Limit Theorem**: mean of sample means, normal distribution
* **Standard error**: std. dev. of the proportion distribution
  * each sample carries a bit of std. error (a part that is not representative of the total population)
  * depends on sample size

# Confidence intervals
Level of confidence of a probability given a particular sample of data

# Hypothesis testing
Is the outcome a result of chance or influenced by outside forces?
