# Data variability

### Range
* Max and min
* Difference between lowest and highest value

### Standard deviation
* Average squared distance from mean
* With balanced data:
  * 68% of data is within 1 std. dev. from the mean 
  * 95% within 2 std. devs.
  * and 99.7% within 3 std. devs.
* Distance more than 1 std. dev. are outliers

### Outliers
* Definition:
  * Abnormal distance from other points
  * Abnormal = ? -> tables, charts, more than 2 std. devs., etc.
* What to do with them?
  * Most likely freekish events that can be ignored
  * Or can show something new, start of a new trend

### Empirical rule
* Also called "Three Sigma Rule" (Sigma = Std. Dev.), or "68-95-99.7 rule"
* Works with symmetrically distributed data (well-centered, symmetrical, bell-shaped)
* Explains that:
  * 68% of data lies within 1 std. dev. of the mean
  * 95% within 2 std. devs.
  * and 99.7% within 3 std. devs.   
* Example:
  * If a data point has a z-score of 1.8, we know that it falls within 95% of the data
  * If a data point has a z-score above 3.0, we know that it's a true outlier because it doesn't fall within the 99.7% of the data

### Types of probability
* **Classical** probability (objective):
  * based on calculations
  * Ex. coin flip
    * we know all possible outcomes
    * all possible outcomes are equally likely to occur   
     
     
* **Empirical** probability (objective):
  * based on calculations
  * Ex: free-throws of basketball
    * variables to take into account (player health, fatigue, game situation...)
    * we can only count of observations made to this point
    * with reliable data from repeatable situations, we can get an idea of what to expect
     
     
* **Subjective** probabilities:
  * not based on calculations but on opinions and knowledge
  * Ex. unemployment rate for this year. Previous years not always reliable, each year is different

### Discrete vs. Continuous random variables

* **random variable**: value of the outcome is unknown
* **discrete**: list/range of possible values (e.g. coin = head | tail), can't be decimals
* **continuous**: can be any value, no limits (range), decimals possible

4. Mean = 1.46 drinks   
  Std. Dev. = 0.94 drinks

  * **Range for 1 std. dev.**:
    * lower limit = 1.46 - 0.94 = 0.52
    * higher limit = 1.46 + 0.94 = 2.40     
    = 0.52 < 1.46 < 2.40 drinks   
    = 1 or 2 drinks ordered = 0.70 + 0.16 (rel. freq.).   
    = **86%** of customers
  
  * **Range for 2 std. devs.**:
    * lower limit = 1.46 - (2 * 0.94) = -0.43
    * higher limit = 1.46 + (2 * 0.94) = 3.35     
    = -0.43 < 1.46 < 3.35 drinks   
    = 0 to 3 drinks ordered = 0.02 + 0.70 + 0.16 + 0.06   
    = **94%** of customers
  
  * **Range for 3 std. devs.**:
    = -1.37 < 1.46 < 4.29 drinks   
    = 0 to 4 drinks ordered = 0.02 + 0.70 + 0.16 + 0.06 + 0.04   
    = **98%** of customers
  
  * 3 std. devs. from mean should capture 99.7% of our data   
    So this single order of 5 drinks is a true outlier
