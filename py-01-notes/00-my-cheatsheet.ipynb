```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def is_leaf(root: TreeNode) -> bool:
    return root.left is None and root.right is None

def is_bst(root: TreeNode) -> bool:
    if not root or is_leaf(root):
        return True
    if (root.left.val < root.val) and (root.right.val > root.val):
        return True and is_bst(root.left) and is_bst(root.right)
    return False

def get_max_bst(root: TreeNode) -> TreeNode:
    if root:
        if is_bst(root):
            if root.left and root.val:
                print('ok', root.val)
                return root
        else:
            get_max_bst(root.left)
            get_max_bst(root.right)
        
def sum_bst(node, s):
    if node:
        s += node.val
        s = sum_bst(node.left, s)
        s = sum_bst(node.right, s)
    return s

bst = TreeNode(1)
bst.left = TreeNode(4, TreeNode(2), TreeNode(4))
bst.right = TreeNode(3, TreeNode(2), TreeNode(5, TreeNode(4), TreeNode(6)))
print(sum_bst(bst, 0))
print(is_bst(bst))
max_bst = get_max_bst(bst)
print(max_bst)
```
