# PATTERN: TWO POINTERS

https://www.educative.io/courses/grokking-the-coding-interview/xlK78P3Xl7E

- Time complexity $O(n)$
- At each step:
  - remove the left element in the window
  - add the next right element to the window

<SPAN style="background:YELLOW;padding: 4px;font-weight: bold;">WHEN TO USE?</SPAN> When dealing with **sorted arrays or linked lists**:     
- In problems where we deal with sorted arrays (or LinkedLists) and need to find a set of elements that fulfill certain constraints, the Two Pointers approach becomes quite useful. The set of elements could be a pair, a triplet or even a subarray.
