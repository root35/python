# Python Execution modes

* **Script** is a file you intend to ***execute*** from the command line, or `python -m` :   
  `__name__` = `__main__`  
  
  
* **Module** is a file you intend to ***import*** from a script or from another module:   
  `__name__` = the file's name without `.py`

```python
print(f'{repr(__name__)}')
```

# Best practices

* Put most code into functions or classes
* Use `__name__` to control execution of your code
* Create a function called `main()` to contain the code you cant to run
* Call other functions from `main()`

```python
def print_hello(data):
    print(f"This is a func saying: {data}")

def main():
    print_hello("Hello world!")

if __name__ == '__main__':
    main()
```
